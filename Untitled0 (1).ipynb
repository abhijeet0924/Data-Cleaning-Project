```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Raw dataset create kar rahe hain simulation ke liye
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-04', None],
    'Customer_ID': [101, 102, 102, 103, 104, 105],
    'Sales': [250, 300, 300, 5000, 150, 200], # 5000 ek outlier hai
    'Profit': [50, 60, 60, 1000, None, 40],
    'Category': ['Electronics', 'Fashion', 'Fashion', 'Electronics', 'Home', 'Home']
}

df = pd.DataFrame(data)
print("--- Raw Data loaded successfully ---")
df.head()

# 1. Duplicates hatana
df.drop_duplicates(inplace=True)

# 2. Missing Values handle karna
df['Date'] = df['Date'].fillna('2023-01-05') # Missing date fill ki
df['Profit'] = df['Profit'].fillna(df['Profit'].median()) # Missing profit mein median dala

# 3. Data type theek karna (Date ko string se datetime mein badalna)
df['Date'] = pd.to_datetime(df['Date'])

print("--- Data Cleaning Done! ---")
df
# IQR Calculate karna
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outlier ko upper bound limit par cap/lock karna
df['Sales'] = np.where(df['Sales'] > upper_bound, upper_bound, df['Sales'])

print("--- Outliers Handled! Final Cleaned Data: ---")
df
# Visualizations plot karna
plt.figure(figsize=(12, 5))

# Plot 1: Sales by Category (Bar Plot)
plt.subplot(1, 2, 1)
sns.barplot(x='Category', y='Sales', data=df, palette='viridis')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')

# Plot 2: Profit Trend Over Time (Line Plot)
plt.subplot(1, 2, 2)
sns.lineplot(x='Date', y='Profit', data=df, marker='o', color='green')
plt.title('Profit Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
